```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
columns = ['sequence_type', 'order', 'samples', 'count']
df = pd.read_csv("rebalance_counters.tsv", sep='\t', header=None, names=columns)
df = df.sort_values(by=['samples', 'order'])
df["order_samples"] = df['order'].astype(str) + "/" + df['samples'].astype(str)

df.set_index('order_samples')
sb.barplot(data=df, x='order_samples', y='count', hue='sequence_type')
plt.xticks(rotation=45)
plt.yscale('log')
plt.xlabel('order/items')
plt.ylabel('rebalance count')
plt.title('Number of re-balances for different\ntree orders and items')
plt.tight_layout()
plt.savefig('img/rebalance_count.png')
```
